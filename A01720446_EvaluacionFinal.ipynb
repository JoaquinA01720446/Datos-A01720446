```python
#Llamamos las librerias a utilizar en el analisis.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Placement_Data_Full_Class.csv')
df.info() #Proporcionamos la informacion general de nuestras variables criticas. 
```

```python
df.head() #Mostramos un muestreo de la base de datos para observar sus valores originales.
```

```python
#Inciamos el analisis. 
plt.figure(figsize=(7,7)) #Seleccionamos el tamaño de la gráfica de barras. 
sns.catplot(x='gender', y='salary', hue='workex', data=df, kind='bar') #Creamos la grafica de barras.
plt.title('Salario ofrecido por genero') #Agregamos un titulo.
#Agregamos un titulo a los valores de x y y.
plt.xlabel('Genero')
plt.ylabel('Salario')
plt.show() #Graficamos.
```

```python
plt.figure(figsize=(6,6)) #Seleccionamos el tamaño de la gráfica de barras.
sns.scatterplot(x='hsc_p', y='degree_p', hue='status', data=df) #Creamos el gráfico de dispersión.
plt.title('Preparatoria vs Universidad') #Agregamos un titulo.
#Agregamos un titulo a los valores de x y y.
plt.xlabel('Preparatoria')
plt.ylabel('Universidad')
plt.show() #Graficamos.
```

```python
sns.set(style='whitegrid') #Agregamos lineas cuadrículas a nuestrao diagrama de caja.
sns.boxplot(x=df['status'], y=df['etest_p']) #Creamos el diagrama de caja.
#Agregamos un titulo a los valores de x y y.
plt.xlabel('Estatus')
plt.ylabel('Test de Empleabilidad')
plt.title('Empleabilidad') #Agregamos un titulo.
plt.show #Graficamos.
```

```python
plt.figure(figsize=(7,7)) #Seleccionamos el tamaño de la gráfica de barras
sns.catplot(x='gender', y='etest_p', hue='status', data=df, kind='bar') #Creamos la gráfica de barras.
plt.title('Probabilidad de conseguir trabajo respecto a test de empleabilidad') #Agregamos un titulo.
#Agregamos un titulo a los valores de x y y.
plt.xlabel('Genero')
plt.ylabel('Promedio de Empleabilidad') #Agregamos un titulo.
plt.show() #Graficamos.
```
