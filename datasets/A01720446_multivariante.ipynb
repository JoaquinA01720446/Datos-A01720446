```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head()
```

```python
men = df.loc[df['sexo']==1]
fem = df.loc[df['sexo']==2]

menweight = men['peso']
femweight = fem['peso']

menwaist = men['cintura']
femwaist = fem['cintura']

sistol = df['sistol']
diastol = df['diastol']

weight = df['peso']
waist = df['cintura']
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(x = sistol, bins = 10)
plt.title('Histograma Sistol')

plt.subplot(1,2,2)
plt.hist(x = diastol, bins = 10)
plt.title('Histograma Diastol')

plt.show()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(x = weight, bins = 10)
plt.title("Histograma Peso")

plt.subplot(1,2,2)
plt.hist(x = waist, bins = 10)
plt.title("Histograma Cintura")

plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x = waist, y = weight)
plt.xlabel("'Cintura'")
plt.ylabel("'Peso'")
plt.title("Cintura x Peso")

plt.show()
```

```python
femweight = fem['peso']
femwaist = fem['cintura']

plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
plt.hist(x = femweight, bins = 10)
plt.title('Peso de las Mujeres')
plt.subplot(1,2,2)
plt.hist(x = femwaist, bins = 10)
plt.title('Cintura de las Mujeres')

plt.show()


menweight = men['peso']
menwaist = men['cintura']

plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
plt.hist(x = menweight, bins = 10)
plt.title('Peso de los Hombres')
plt.subplot(1,2,2)
plt.hist(x = menwaist, bins = 10)
plt.title('Cintura de los Hombres')

plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.scatter(x = femwaist, y = femweight)
plt.xlabel('Cintura Fem')
plt.ylabel('Peso Fem')
plt.title('Cintura x Peso Mujeres')

plt.show()

plt.figure(figsize = (6,6))
plt.scatter(x = menwaist, y = menweight)
plt.xlabel('Cintura Men')
plt.ylabel('Peso Men')
plt.title('Cintura x Peso Hombres')

plt.show()
```
